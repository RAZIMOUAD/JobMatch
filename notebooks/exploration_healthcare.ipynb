```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configuration des graphiques
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("=" * 80)
print("üè• EXPLORATION APPROFONDIE - DATASET HEALTHCARE FINAL")
print("=" * 80)

# Chargement
file_path = '../data/processed/healthcare_jobs_final.csv'
df_hc = pd.read_csv(file_path)

print(f"\n‚úÖ Dataset charg√© avec succ√®s !")
print(f"\nüìä INFORMATIONS G√âN√âRALES")
print("=" * 80)
print(f"Nombre d'offres      : {len(df_hc):,}")
print(f"Nombre de colonnes   : {len(df_hc.columns)}")
print(f"Taille m√©moire       : {df_hc.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

print(f"\nüìù COLONNES DISPONIBLES :")
for i, col in enumerate(df_hc.columns, 1):
    dtype = df_hc[col].dtype
    non_null = df_hc[col].notna().sum()
    null_pct = (df_hc[col].isna().sum() / len(df_hc)) * 100
    print(f"  {i}. {col:30s} | Type: {str(dtype):10s} | Rempli: {non_null:6,} ({100-null_pct:5.1f}%)")

print(f"\nüìã APER√áU DES 30 PREMI√àRES LIGNES :")
print("=" * 80)
display(df_hc.head(30))

print(f"\nüìã TYPES DE DONN√âES D√âTAILL√âS :")
print(df_hc.dtypes)

print(f"\nüîç STATISTIQUES DESCRIPTIVES (colonnes num√©riques) :")
print(df_hc.describe())

print("\n" + "=" * 80)
print("‚úÖ CELLULE 1 TERMIN√âE - Vue d'ensemble charg√©e")
print("=" * 80)
```

```python
print("\n" + "=" * 80)
print("TEST 1 : ANALYSE DES VALEURS MANQUANTES")
print("=" * 80)

# Compter les valeurs manquantes
missing_data = pd.DataFrame({
    'Colonne': df_hc.columns,
    'Valeurs manquantes': df_hc.isnull().sum().values,
    'Pourcentage': (df_hc.isnull().sum().values / len(df_hc) * 100)
})

missing_data = missing_data.sort_values('Valeurs manquantes', ascending=False)

print("\nüìä VALEURS MANQUANTES PAR COLONNE :")
print("-" * 80)
for idx, row in missing_data.iterrows():
    if row['Valeurs manquantes'] > 0:
        print(f"‚ö†Ô∏è  {row['Colonne']:30s} : {row['Valeurs manquantes']:6.0f} ({row['Pourcentage']:5.2f}%)")
    else:
        print(f"‚úÖ  {row['Colonne']:30s} : {row['Valeurs manquantes']:6.0f} ({row['Pourcentage']:5.2f}%)")

# Si formatted_experience_level a des NaN (attendu)
if df_hc['formatted_experience_level'].isna().sum() > 0:
    print(f"\nüìä D√âTAIL : formatted_experience_level")
    print("-" * 80)
    exp_missing = df_hc['formatted_experience_level'].isna().sum()
    exp_present = df_hc['formatted_experience_level'].notna().sum()
    print(f"   Manquantes  : {exp_missing:6,} ({exp_missing/len(df_hc)*100:5.2f}%)")
    print(f"   Pr√©sentes   : {exp_present:6,} ({exp_present/len(df_hc)*100:5.2f}%)")
    
    if exp_present > 0:
        print(f"\n   Distribution des valeurs pr√©sentes :")
        exp_dist = df_hc['formatted_experience_level'].value_counts()
        for level, count in exp_dist.items():
            print(f"      {str(level):25s} : {count:6,} ({count/exp_present*100:5.2f}%)")

# Visualisation
fig, ax = plt.subplots(figsize=(10, 6))
missing_data_to_plot = missing_data[missing_data['Valeurs manquantes'] > 0]

if len(missing_data_to_plot) > 0:
    ax.barh(missing_data_to_plot['Colonne'], missing_data_to_plot['Pourcentage'], color='coral')
    ax.set_xlabel('Pourcentage de valeurs manquantes')
    ax.set_title('Valeurs manquantes par colonne')
    ax.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print("\n‚úÖ Aucune valeur manquante √† visualiser !")

print("\n" + "=" * 80)
```

```python
print("\n" + "=" * 80)
print("TEST 2 : R√âPARTITION PAR NIVEAU DE CONFIANCE")
print("=" * 80)

if 'confidence_level' in df_hc.columns:
    conf_dist = df_hc['confidence_level'].value_counts()
    
    print(f"\nüìä DISTRIBUTION :")
    print("-" * 80)
    for level, count in conf_dist.items():
        pct = (count / len(df_hc)) * 100
        emoji = "üü¢" if level == "High" else "üü°" if level == "Medium" else "üîµ"
        print(f"{emoji} {level:10s} : {count:6,} offres ({pct:5.2f}%)")
    
    # Exemples par niveau
    print(f"\nüìã EXEMPLES PAR NIVEAU DE CONFIANCE :")
    print("=" * 80)
    
    for level in ['High', 'Medium', 'Support']:
        if level in conf_dist.index:
            level_data = df_hc[df_hc['confidence_level'] == level]
            samples = level_data.sample(min(3, len(level_data)), random_state=42)
            
            print(f"\n{level.upper()} CONFIDENCE ({len(level_data):,} offres) :")
            print("-" * 80)
            for i, (idx, row) in enumerate(samples.iterrows(), 1):
                print(f"\n   Exemple {i}:")
                print(f"   Job ID       : {row['job_id']}")
                print(f"   Title        : {row['title']}")
                print(f"   Location     : {row['location']}")
                print(f"   Experience   : {row['formatted_experience_level']}")
                print(f"   Skill_name   : {row['skill_name']}")
    
    # Visualisation
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Graph 1: Bar chart
    colors_map = {'High': 'green', 'Medium': 'orange', 'Support': 'steelblue', 'Unknown': 'gray'}
    colors = [colors_map.get(x, 'gray') for x in conf_dist.index]
    
    axes[0].bar(conf_dist.index, conf_dist.values, color=colors, edgecolor='black')
    axes[0].set_ylabel('Nombre d\'offres')
    axes[0].set_title('R√©partition par niveau de confiance')
    axes[0].grid(axis='y', alpha=0.3)
    
    for i, (level, count) in enumerate(conf_dist.items()):
        axes[0].text(i, count + 200, f'{count:,}', ha='center', fontweight='bold')
    
    # Graph 2: Pie chart
    axes[1].pie(conf_dist.values, labels=conf_dist.index, autopct='%1.1f%%', 
                colors=colors, startangle=90)
    axes[1].set_title('Proportion par niveau de confiance')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("‚ö†Ô∏è Colonne 'confidence_level' non trouv√©e")

print("\n" + "=" * 80)
```

```python
print("\n" + "=" * 80)
print("TEST 3 : ANALYSE G√âOGRAPHIQUE D√âTAILL√âE")
print("=" * 80)

print(f"\nüåç STATISTIQUES GLOBALES :")
print("-" * 80)
print(f"Locations uniques           : {df_hc['location'].nunique():,}")
print(f"Total offres                : {len(df_hc):,}")
print(f"Moyenne offres par location : {len(df_hc) / df_hc['location'].nunique():.1f}")

# Distribution
loc_counts = df_hc['location'].value_counts()

print(f"\nüìä CONCENTRATION G√âOGRAPHIQUE :")
print("-" * 80)
top_5_count = loc_counts.head(5).sum()
top_10_count = loc_counts.head(10).sum()
top_20_count = loc_counts.head(20).sum()
top_50_count = loc_counts.head(50).sum()

print(f"Top 5 locations   : {top_5_count:6,} offres ({top_5_count/len(df_hc)*100:5.2f}%)")
print(f"Top 10 locations  : {top_10_count:6,} offres ({top_10_count/len(df_hc)*100:5.2f}%)")
print(f"Top 20 locations  : {top_20_count:6,} offres ({top_20_count/len(df_hc)*100:5.2f}%)")
print(f"Top 50 locations  : {top_50_count:6,} offres ({top_50_count/len(df_hc)*100:5.2f}%)")

print(f"\nüèÜ TOP 30 LOCATIONS :")
print("-" * 80)
for i, (loc, count) in enumerate(loc_counts.head(30).items(), 1):
    pct = (count / len(df_hc)) * 100
    bar = "‚ñà" * int(pct * 2)  # Barre visuelle
    print(f"{i:2d}. {loc:40s} : {count:5,} ({pct:5.2f}%) {bar}")

print(f"\nüèöÔ∏è LOCATIONS RARES (moins de 5 offres) :")
print("-" * 80)
rare_locs = loc_counts[loc_counts < 5]
print(f"Nombre de locations rares : {len(rare_locs):,} ({len(rare_locs)/df_hc['location'].nunique()*100:.2f}%)")
print(f"Total offres dans ces locations : {rare_locs.sum():,} ({rare_locs.sum()/len(df_hc)*100:.2f}%)")

# Visualisation
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Graph 1: Top 20 locations
loc_counts.head(20).plot(kind='barh', ax=axes[0, 0], color='mediumseagreen')
axes[0, 0].set_xlabel('Nombre d\'offres')
axes[0, 0].set_title('Top 20 Locations Healthcare')
axes[0, 0].invert_yaxis()

# Graph 2: Distribution par taille de ville
bins = [0, 10, 50, 100, 500, 10000]
labels = ['1-10', '11-50', '51-100', '101-500', '500+']
loc_size_dist = pd.cut(loc_counts, bins=bins, labels=labels).value_counts().sort_index()

axes[0, 1].bar(range(len(loc_size_dist)), loc_size_dist.values, color='steelblue', edgecolor='black')
axes[0, 1].set_xticks(range(len(loc_size_dist)))
axes[0, 1].set_xticklabels(loc_size_dist.index, rotation=0)
axes[0, 1].set_ylabel('Nombre de locations')
axes[0, 1].set_xlabel('Nombre d\'offres par location')
axes[0, 1].set_title('Distribution des tailles de locations')
axes[0, 1].grid(axis='y', alpha=0.3)

# Graph 3: Courbe cumulative
cumsum = (loc_counts.cumsum() / loc_counts.sum() * 100)
axes[1, 0].plot(range(len(cumsum)), cumsum.values, linewidth=2, color='darkgreen')
axes[1, 0].fill_between(range(len(cumsum)), cumsum.values, alpha=0.3, color='green')
axes[1, 0].set_xlabel('Nombre de locations (class√©es)')
axes[1, 0].set_ylabel('% cumul√© d\'offres')
axes[1, 0].set_title('Courbe cumulative - Concentration g√©ographique')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].axhline(y=50, color='red', linestyle='--', label='50%')
axes[1, 0].axhline(y=80, color='orange', linestyle='--', label='80%')
axes[1, 0].legend()

# Graph 4: Locations rares vs communes
categories = ['Top 50\nlocations', 'Locations\nrares (<5)']
counts = [top_50_count, rare_locs.sum()]
colors_cat = ['green', 'red']

axes[1, 1].bar(categories, counts, color=colors_cat, edgecolor='black')
axes[1, 1].set_ylabel('Nombre d\'offres')
axes[1, 1].set_title('Concentration vs Dispersion')
axes[1, 1].grid(axis='y', alpha=0.3)

for i, count in enumerate(counts):
    pct = (count / len(df_hc)) * 100
    axes[1, 1].text(i, count + 200, f'{count:,}\n({pct:.1f}%)', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
```

```python
print("\n" + "=" * 80)
print("TEST 4 : ANALYSE DES TITRES - Patterns Healthcare")
print("=" * 80)

# Mots dans les titres
from collections import Counter
import re

# Nettoyer et splitter les titres
all_words = []
for title in df_hc['title'].dropna():
    # Nettoyer (enlever ponctuation, minuscules)
    words = re.findall(r'\b[a-zA-Z]{3,}\b', title.lower())
    all_words.extend(words)

word_counts = Counter(all_words)

print(f"\nüìä STATISTIQUES SUR LES TITRES :")
print("-" * 80)
print(f"Titres uniques            : {df_hc['title'].nunique():,}")
print(f"Mots uniques dans titres  : {len(word_counts):,}")
print(f"Mots totaux               : {len(all_words):,}")

print(f"\nüîç TOP 50 MOTS LES PLUS FR√âQUENTS :")
print("-" * 80)
for i, (word, count) in enumerate(word_counts.most_common(50), 1):
    pct = (count / len(df_hc)) * 100
    print(f"{i:2d}. {word:20s} : {count:6,} ({pct:5.2f}% des offres)")

# Cat√©gories de jobs
print(f"\nüìä CAT√âGORIES DE JOBS (par mots-cl√©s dans titre) :")
print("-" * 80)

categories = {
    'Nurses (RN/LPN/CNA)': ['nurse', 'nursing', 'lpn', 'cna', 'rn'],
    'Therapists': ['therapist', 'therapy', 'physical', 'occupational', 'speech'],
    'Medical Assistants': ['medical assistant', 'assistant'],
    'Physicians/Doctors': ['physician', 'doctor', 'md'],
    'Dental': ['dental', 'dentist', 'hygienist'],
    'Pharmacy': ['pharmacy', 'pharmacist', 'pharmaceutical'],
    'Mental Health': ['mental', 'counselor', 'psychologist', 'psychiatric'],
    'Administrative': ['manager', 'coordinator', 'administrator', 'director'],
    'Technical/Support': ['technician', 'tech', 'analyst', 'specialist']
}

for category, keywords in categories.items():
    pattern = '|'.join(keywords)
    count = df_hc['title'].str.lower().str.contains(pattern, regex=True, na=False).sum()
    pct = (count / len(df_hc)) * 100
    print(f"{category:30s} : {count:6,} ({pct:5.2f}%)")

# Visualisation
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Graph 1: Top 30 mots
top_words = dict(word_counts.most_common(30))
axes[0, 0].barh(list(top_words.keys())[::-1], list(top_words.values())[::-1], color='mediumseagreen')
axes[0, 0].set_xlabel('Fr√©quence')
axes[0, 0].set_title('Top 30 mots dans les titres')

# Graph 2: Cat√©gories de jobs
cat_counts = {cat: df_hc['title'].str.lower().str.contains('|'.join(kw), regex=True, na=False).sum() 
              for cat, kw in categories.items()}
axes[0, 1].bar(range(len(cat_counts)), list(cat_counts.values()), color='steelblue', edgecolor='black')
axes[0, 1].set_xticks(range(len(cat_counts)))
axes[0, 1].set_xticklabels(list(cat_counts.keys()), rotation=45, ha='right')
axes[0, 1].set_ylabel('Nombre d\'offres')
axes[0, 1].set_title('Distribution par cat√©gorie de job')
axes[0, 1].grid(axis='y', alpha=0.3)

# Graph 3: Longueur des titres
title_lengths = df_hc['title'].str.len()
axes[1, 0].hist(title_lengths, bins=50, color='coral', edgecolor='black')
axes[1, 0].axvline(title_lengths.mean(), color='red', linestyle='--', linewidth=2, 
                   label=f'Moyenne: {title_lengths.mean():.1f}')
axes[1, 0].set_xlabel('Nombre de caract√®res')
axes[1, 0].set_ylabel('Fr√©quence')
axes[1, 0].set_title('Distribution longueur des titres')
axes[1, 0].legend()
axes[1, 0].grid(axis='y', alpha=0.3)

# Graph 4: Nombre de mots par titre
title_word_counts = df_hc['title'].str.split().str.len()
axes[1, 1].hist(title_word_counts, bins=20, color='purple', edgecolor='black')
axes[1, 1].axvline(title_word_counts.mean(), color='red', linestyle='--', linewidth=2,
                   label=f'Moyenne: {title_word_counts.mean():.1f}')
axes[1, 1].set_xlabel('Nombre de mots')
axes[1, 1].set_ylabel('Fr√©quence')
axes[1, 1].set_title('Distribution nombre de mots par titre')
axes[1, 1].legend()
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
```

```python
print("\n" + "=" * 80)
print("üßπ NETTOYAGE : Offres sans skill_name")
print("=" * 80)

missing_skill = df_hc[df_hc['skill_name'].isna()].copy()

# V√©rifier si elles ont des keywords Healthcare dans description
healthcare_keywords = [
    'nurse', 'patient', 'medical', 'health', 'clinical', 'hospital',
    'doctor', 'physician', 'therapy', 'care provider', 'healthcare',
    'registered nurse', 'lpn', 'rn', 'cna', 'practitioner', 'treatment'
]

pattern = '|'.join(healthcare_keywords)

# Chercher dans description (plus fiable que titre)
missing_skill['has_hc_in_desc'] = missing_skill['description'].astype(str).str.lower().str.contains(
    pattern, regex=True, na=False
)

# Statistiques
has_hc = missing_skill[missing_skill['has_hc_in_desc'] == True]
no_hc = missing_skill[missing_skill['has_hc_in_desc'] == False]

print(f"\nüìä R√âSULTATS DU SCAN :")
print(f"   Avec keywords Healthcare dans description : {len(has_hc)} ({len(has_hc)/len(missing_skill)*100:.1f}%)")
print(f"   SANS keywords Healthcare                  : {len(no_hc)} ({len(no_hc)/len(missing_skill)*100:.1f}%)")

print(f"\nüìã EXEMPLES √Ä SUPPRIMER (sans keywords HC) :")
print("-" * 80)
samples_no_hc = no_hc.sample(min(10, len(no_hc)), random_state=42)
for i, (idx, row) in enumerate(samples_no_hc.iterrows(), 1):
    print(f"\n   #{i}")
    print(f"   Title : {row['title']}")
    print(f"   Desc  : {str(row['description'])[:150]}...")

print(f"\nüí° D√âCISION :")
print(f"   ‚úÖ GARDER : {len(has_hc)} offres (ont keywords Healthcare dans description)")
print(f"   ‚ùå SUPPRIMER : {len(no_hc)} offres (probablement faux positifs)")

# Cr√©er le dataset nettoy√©
df_hc_cleaned = df_hc[
    (df_hc['skill_name'].notna()) |  # Garder ceux qui ont skill_name
    ((df_hc['skill_name'].isna()) & (df_hc['job_id'].isin(has_hc['job_id'])))  # Garder ceux sans skill_name MAIS avec keywords HC
].copy()

print(f"\nüìä DATASET NETTOY√â :")
print(f"   Avant  : {len(df_hc):,} offres")
print(f"   Apr√®s  : {len(df_hc_cleaned):,} offres")
print(f"   Supprim√© : {len(df_hc) - len(df_hc_cleaned):,} offres")
```

```python
print("\n" + "=" * 80)
print("üßπ NETTOYAGE FINAL : Attorneys non-Healthcare")
print("=" * 80)

# Identifier tous les attorneys
attorneys = df_hc_cleaned[df_hc_cleaned['title'].str.lower().str.contains('attorney|lawyer', regex=True, na=False)]

print(f"\nüìä Total attorneys/lawyers : {len(attorneys)}")

# Identifier ceux qui sont VRAIMENT healthcare
healthcare_law_keywords = [
    'medical', 'health', 'malpractice', 'elder', 'nursing home',
    'hospital', 'patient', 'physician', 'medicare', 'medicaid'
]

pattern_hc_law = '|'.join(healthcare_law_keywords)

# Chercher dans TITRE (plus fiable)
attorneys['is_hc_law'] = attorneys['title'].str.lower().str.contains(pattern_hc_law, regex=True, na=False)

# Si pas dans titre, chercher dans description
attorneys['is_hc_law_desc'] = attorneys['description'].astype(str).str.lower().str.contains(
    'medical malpractice|healthcare law|elder law|nursing home litigation',
    regex=True,
    na=False
)

# Combiner
healthcare_attorneys = attorneys[(attorneys['is_hc_law']) | (attorneys['is_hc_law_desc'])]
non_healthcare_attorneys = attorneys[~((attorneys['is_hc_law']) | (attorneys['is_hc_law_desc']))]

print(f"\nüìä R√âPARTITION :")
print(f"   ‚úÖ Healthcare attorneys  : {len(healthcare_attorneys):4d} ({len(healthcare_attorneys)/len(attorneys)*100:5.1f}%)")
print(f"   ‚ùå Non-Healthcare        : {len(non_healthcare_attorneys):4d} ({len(non_healthcare_attorneys)/len(attorneys)*100:5.1f}%)")

print(f"\nüìã EXEMPLES √Ä GARDER (Healthcare law) :")
if len(healthcare_attorneys) > 0:
    for idx, row in healthcare_attorneys.head(10).iterrows():
        print(f"   ‚úÖ {row['title']}")

print(f"\nüìã EXEMPLES √Ä SUPPRIMER (Non-Healthcare) :")
for idx, row in non_healthcare_attorneys.sample(min(15, len(non_healthcare_attorneys)), random_state=42).iterrows():
    print(f"   ‚ùå {row['title']}")

# Dataset final sans les attorneys non-HC
df_final = df_hc_cleaned[
    ~df_hc_cleaned['job_id'].isin(non_healthcare_attorneys['job_id'])
].copy()

print(f"\nüìä DATASET FINAL :")
print(f"   Avant suppression attorneys : {len(df_hc_cleaned):,}")
print(f"   Apr√®s suppression           : {len(df_final):,}")
print(f"   Supprim√©                    : {len(df_hc_cleaned) - len(df_final):,}")

print("\n" + "=" * 80)
```
